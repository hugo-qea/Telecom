## <span style="color:green"> Part I : Theory </span>

### <span style="color:red"> Orthogonal Polynomials </span>

#### <span style="color:DodgerBlue"> 1. </span> Show that $P_n$ changes sign at least once in $\mathcal{X}$.

We consider the constructed orthogonal family $(P_n)_{n \in \mathbb{N}}$ through Gram-Schmidt process from $(X^n)_{n \in N}$.

First, we have $\tilde{P}_0 = 1$ (it respects the normalization condition) since it is the first term of the process.
Then, according to the process, we have $$\forall n \in \mathbb{N}^* : {P}_n =  X^n - Q_n$$ with $Q_n \in Span((P_k)_{k < n}) = \mathbb{R}_{n-1}[X]$.

Thus by operations on the degrees, we have : $$\forall n \in \mathbb{N}, \deg(\tilde{P}_n) = n$$

Take $n \in \mathbb{N}^*$, let's suppose by absurd that $P_n$ never changes sign over $\mathcal{X}$. Let's suppose that $\forall x \in \mathcal{X}, P_n(x) \geq 0$ (otherwise, same reasoning but with minuses).

However, by orthogonality, we have : $$ \langle P_0, P_n \rangle = 0$$

Thus, $$ \int_{\mathcal{X}} \omega P_n = 0$$ and since $\omega$ is non negative, we have $P_n(x) \omega(x) = 0$ almost everywhere, and since $\omega$ is a non-negative weight function there exists $A \subset \mathcal{X}$ with a non-zero Lebesgue measure such that $\forall x \in A, \omega(x) > 0$.

Therefore, $P_n$ has a infinite number of zeros (every $x \in A$ to begin with), thus $P_n = 0$, which is a contradiction since $P_n$ is a polynomial of degree $n$.

Finally, we have proved that $P_n$ changes sign at least once in $\mathcal{X}$.

#### <span style="color:DodgerBlue"> 2. </span> Let $y_1,...,y_j$ the only points in $\mathcal{X}$ where $P_n$ changes its sign. Show that j = n and conclude.

Once again let's resolve by absurd.

Let's suppose that $j \neq n$. Then, we have $j < n$.

Since $P_n$ changes sign on $\mathcal{X}$ only each $y_i$, we have : $ P_nQ(x) \leq 0 \forall x \in \mathcal{X}$ with $Q = \prod_{i=1}^j (X-y_i)$.

Indeed, since $P_n$ changes sign on $\mathcal{X}$ only each $y_i$, each y_i have odd multiplicity and the potential other roots in $\mathcal{X}$ have even multiplicity.

As a result, we have : $$\int_{\mathcal{X}} \omega P_nQ > 0$$

However, by orthogonality we have that $P_n$ orthogonal to $\mathbb{R}_{n-1}[X]$ and $ Q \in \mathbb{R}_{j}[X] \subset \mathbb{R}_{n-1}[X]$, thus : $$\int_{\mathcal{X}} \omega P_nQ = 0$$

Therefore we have a contradiction and we can conclude that $j = n$.

Finally, we have proved that $P_n$ changes sign on $n$ distinct points in $\mathcal{X}$. However, $P_n$ is of degree $n$, thus it has at $n$ roots.

Thus, $P_n$ has $n$ distinct simple roots in $\mathcal{X}$.

### <span style="color:red"> Necessary conditions </span>

#### <span style="color:DodgerBlue"> 3. </span> Using a particular choice for $f$, show that the nodes $(x_i)_i$ must be roots of $P_{n+1}$.

As we saw earlier, $$\forall k \leq n, \int_{\mathcal{X}}  P_{n+1}(x) x^k \omega(x) dx = 0$$

Thus, $$\forall k \leq n,\sum\limits_{i=0}^{n} \omega_i P_{n+1}(x_i) x_i^k = 0$$

If we consider the Vandermonde matrix $V$ of the points $x_i$ and the vector $b$ of the coefficients $\omega_i P_{n+1}(x_i)$, we have : $$V^Tb = 0$$

However, we have $x_0 < x_1 < ... < x_n$ since they are the nodes, thus $V$ is invertible and $b=0$.

Finally, we have $\forall i \in \{0,...,n\}, w_i P_{n+1}(x_i) = 0$. But, $\forall i \in \{0,...,n\}, w_i > 0$ and the $x_i$ are distinct so they are the $n+1$ roots of $P_{n+1}$.


#### <span style="color:DodgerBlue"> 4. </span> Conclude the uniqueness of $(x_i)_i$ and $(w_i)_i$.

We have seen that the $x_i$ are unique.

To determine the weights :
Consider the Lagrange basis associated to the nodes : $$ \forall i \in \{0,...,n\}, l_i(x) = \prod_{j \neq i} \frac{x-x_j}{x_i-x_j}$$

We notice that $l_j$ is of degree n. Thus its gaussian quadrature is perfect : $$\int_{\mathcal{X}} l_j(x) \omega(x) dx = \sum\limits_{i=0}^{n} \omega_i l_j(x_i) = \sum\limits_{i=0}^{n} \omega_i \delta_{ij} = w_j$$

We thus have found a unique formula for the weights : $$w_j = \int_{\mathcal{X}} l_j(x) \omega(x) dx$$


### <span style="color:red"> Sufficient conditions </span>

#### <span style="color:DodgerBlue"> 5. </span> Using Euclidean division, show that the necessary conditions are sufficient for the rule to be of order at least $2n + 1$.

Take $P \in \mathbb{R}_{2n+1}[X]$ and its euclidean division : $$ P = P_{n+1}Q + R$$ with $R \in \mathbb{R}_{n}[X]$

We have : $$\int_{\mathcal{X}} P(x) \omega(x) dx = \int_{\mathcal{X}} P_{n+1}(x) Q(x) \omega(x) dx + \int_{\mathcal{X}} R(x) \omega(x) dx$$

As before, since Q is at most of degree $n$, we have : $$\int_{\mathcal{X}} P_{n+1}(x) Q(x) \omega(x) dx = 0$$

Thus, $$\int_{\mathcal{X}} P(x) \omega(x) dx = \int_{\mathcal{X}} R(x) \omega(x) dx$$

Then since $R$ is of degree $n$ or less, we can decompose it over the Lagrange basis : $$R(x) = \sum\limits_{i=0}^{n} R(x_i) l_i(x)$$

Thus, $$\int_{\mathcal{X}} P(x) \omega(x) dx = \sum\limits_{i=0}^{n} R(x_i) \int_{\mathcal{X}} l_i(x) \omega(x) dx = \sum\limits_{i=0}^{n} R(x_i) w_i$$

But, since the $x_i$ are the roots of $P_{n+1}$, we have : $$\int_{\mathcal{X}} P(x) \omega(x) dx = \sum\limits_{i=0}^{n} (R+ P_{n+1}Q)(x_i) w_i = \sum\limits_{i=0}^{n} P(x_i) w_i$$

Thus, we have proved that the gaussian quadrature is exact for polynomials of degree $2n+1$ or less. It therefore of order at least $2n+1$.

#### <span style="color:DodgerBlue"> 6. </span> Using a specific choice of $f$ , show that the order of the rule cannot be larger than $2n + 1$.

Take $P = P_{n+1}^2$. We clearly have that $P$ is of degree $2n+2$ and we have : $ \int_\mathcal{X} P(x)\omega(x) dx > 0$.

However, we have : $$\sum\limits_{i=0}^{n} \omega_i P(x_i) = \sum\limits_{i=0}^{n} \omega_i P_{n+1}(x_i)^2 = 0$$

Thus the gaussian quadrature is not exact for P of degree $2n+2$. Therefore, the gaussian quadrature is of order $2n+1$.

#### <span style="color:DodgerBlue"> 7. </span> Find $P_0$, $P_1$ and $P_2$. Deduce the corresponding Gauss-Legendre quadrature for $n = 1$.

#### Gauss-Legendre quadrature

$P_0 = 1$

$P_1 = X - \int_{-1}^{1} \frac{x}{2}dx = X$


$P_2 = X^2 - (\int_{-1}^{1} \frac{3x^3}{4}dx) X - \int_{-1}^{1} \frac{x^2}{2} dx = X^2 - \frac{1}{3}$ However, we want the polynomials to satisfy the normalization condition : $P_2(1) = 1$

Thus, we have : $P_2 = \frac{3}{2} (X^2 -  \frac{1}{3})$

The nodes of the corresponding Gauss-Legendre quadrature are thus $\frac{1}{\sqrt{3}}$ and $-\frac{1}{\sqrt{3}}$. The

The associated weights : $$w_0 = \int_{-1}^{1} \frac{1}{2} \frac{x -\frac{1}{\sqrt{3}}}{- \frac{1}{\sqrt{3}} - \frac{1}{\sqrt{3}}} dx = \int_{-1}^{1} \frac{\sqrt{3}}{4}(\frac{1}{\sqrt{3}} -x) dx = \frac{1}{2}$$

$$w_1 = \int_{-1}^{1} \frac{1}{2} \frac{x +\frac{1}{\sqrt{3}}}{\frac{1}{\sqrt{3}} + \frac{1}{\sqrt{3}}} dx = \int_{-1}^{1} \frac{\sqrt{3}}{4}(x + \frac{1}{\sqrt{3}}) dx = \frac{1}{2}$$

Finally, $$\hat{I}(f) = \frac{1}{2}(f(\frac{1}{\sqrt{3}}) + f(-\frac{1}{\sqrt{3}}))$$

#### Hermite polynomials

$P_0 = 1$

$P_1 = X - \int_{-\infty}^{\infty} x \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} dx = X$


$P_2$ = $X^2 - (\frac{\int_{-\infty}^{\infty} x^3 \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} dx}{\int_{-\infty}^{\infty} x^2 \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} dx}) X - \int_{-\infty}^{\infty} x^2 \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} dx = X^2 - 1$

Thus, the nodes of the corresponding Gauss-Hermite quadrature are $-1$ and $1$.

And the associated weights : $$w_0 = \int_{-\infty}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \frac{x -1}{-1 - 1} dx = \int_{-\infty}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \frac{1}{2} dx = \sqrt{\frac{\pi}{2}}$$

$$w_1 = \int_{-\infty}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \frac{x +1}{1 + 1} dx = \int_{-\infty}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \frac{1}{2} dx = \sqrt{\frac{\pi}{2}}$$

Finally, $$\hat{I}(f) = \sqrt{\frac{\pi}{2}}(f(1) + f(-1))$$

#### Laguerre polynomials

$P_0 = 1$

$P_1 = X - \int_{0}^{\infty} x e^{-x} dx = X - 1$

$P_2 = X^2 - \frac{\int_{0}^{\infty} x^2(x-1) e^{-x} dx}{\int_{0}^{\infty} (x-1)^2 e^{-x} dx} (X-1) - \int_{0}^{\infty} x^2 e^{-x} dx = X^2 - (\int_{0}^{\infty}2x^2 e^{-x} dx)(X-1) - \int_{0}^{\infty} x^2 e^{-x} dx = X^2 - 4X + 4 - 2$ = $X^2 - 4X + 2$

Thus, the nodes of the corresponding Gauss-Laguerre quadrature are $2 - \sqrt{2}$ and $2 + \sqrt{2}$.

And the associated weights : $$w_0 = \int_{0}^{\infty} e^{-x} \frac{x - (2 + \sqrt{2})}{(2 - \sqrt{2}) - (2 + \sqrt{2})} dx = \int_{0}^{\infty} e^{-x} \frac{x - (2 + \sqrt{2})}{-2 \sqrt{2}} dx = \frac{-1}{2 \sqrt{2}} + \frac{2 + \sqrt{2}}{2 \sqrt{2}} = \frac{2 + \sqrt{2}}{4}$$

$$w_1 = \int_{0}^{\infty} e^{-x} \frac{x - (2 - \sqrt{2})}{(2 + \sqrt{2}) - (2 - \sqrt{2})} dx = \int_{0}^{\infty} e^{-x} \frac{x - (2 - \sqrt{2})}{2 \sqrt{2}} dx = \frac{1}{2 \sqrt{2}} + \frac{-2 + \sqrt{2}}{2 \sqrt{2}} = \frac{2 - \sqrt{2}}{4}$$

Finally, $$\hat{I}(f) = \frac{2 + \sqrt{2}}{4} f(2 - \sqrt{2}) + \frac{2 - \sqrt{2}}{4} f(2 + \sqrt{2})$$

#### <span style="color:DodgerBlue"> 9. </span> Implement a function def gauss_legendre_simple(f, a, b, nodes, wieghts) that approximates the integral of a function $f$ over $[a, b]$ using the Gaussian-Quadrature rule. Test your implementation on some analytically integrable function of your choice.

```python
def gauss_legendre_simple(f, a, b, nodes, weights) :
    degree = len(nodes)
    I_hat = 0
    m = (a + b) / 2
    h = (b - a) / 2
    for i in range(degree) :
        I_hat += weights[i] * f(m + h * nodes[i])
    I_hat = h*I_hat
    return I_hat
```

#### <span style="color:DodgerBlue"> 10. </span> Implement a function def gauss_legendre(f, n, M, a, b) that performs the composite rule of Gauss-Legendre over $M$ sub-intervals of $[a, b]$.

```python
def gauss_legendre(f, n, M, a, b):
    I_hat = 0
    b_i = a
    nodes, weights = roots_legendre(n)
    for i in range(M):
        a_i = b_i
        b_i = a_i + (b - a) / M
        I_hat += gauss_legendre_simple(f, a_i, b_i, nodes, weights)
    return I_hat
```

```python
#Testing the method

a = 0
b = np.pi
n = 100
M = 10
print(gauss_legendre(f1, n, M, a, b))
print(quad(f1, a, b)[0])
```

#### <span style="color:DodgerBlue"> 11. </span> Write an identical function but with a vectorized implementation gauss_legendre_vectorized(f, n, M, a, b) with numpy (does not contain any loop or list comprehensions).

```python
def gauss_legendre_vectorized(f, n, M, a, b):
    nodes, weights = roots_legendre(n)
    h = (b - a) / M
    x_start = np.linspace(a, b, M, endpoint=False)
    x_end = x_start + h
    m_i = (x_start + x_end) / 2
    m_i = m_i[:, None] + (h / 2) * nodes
    
    weights = (h / 2) * weights
    
    return np.sum(weights * f(m_i), axis=1).sum()
```

```python
#Testing the method

a = 0
b = np.pi
n = 100
M = 10
print(gauss_legendre_vectorized(f1, n, M, a, b))
print(quad(f1, a, b)[0])
```

#### <span style="color:DodgerBlue"> 13. </span> Write a function def trapezoidal that computes the composite trapezoidal rule of Ex 1. Compare the integration error with the Gauss-Legendre method with the same values of M and the number of nodes. What do you conclude?


```python
def trapezoidal(f, a, b, M):
    h = (b - a) / M
    x = np.linspace(a, b, M+1)
    y = f(x)
    integral = 0
    for i in range(len(x)) :
        if (i != 0) & (i != M) :
            integral += h*y[i]
        else :
            integral += h*y[i]/2
    return integral
```

```python
#Testing the method

a = 0
b = np.pi
M = 10
print(trapezoidal(f1, a, b, M))
print(quad(f1, a, b)[0])
```

```python
# Trapezoidal rule vs Gauss-Legendre

exact = 2

M = np.arange(10, 1000)
trapezoidal_errors = np.zeros(len(M))
gauss_legendre_errors = np.zeros(len(M))

for k in range(len(M)):
    nodes , weights = roots_legendre(M[k])
    trapezoidal_errors[k] = abs(trapezoidal(f1, 0, np.pi, M[k]) - exact)
    gauss_legendre_errors[k] = abs(gauss_legendre_simple(f1, 0, np.pi, nodes, weights) - exact)
plt.plot(M, trapezoidal_errors, label='Trapezoidal')
plt.plot(M, gauss_legendre_errors, label='Gauss-Legendre')
plt.xlabel('M')
plt.ylabel('Error')
plt.yscale('log')
plt.legend()
plt.show()
```

The Gauss quadrature method is more accurate than the trapezoidal method. This is because the trapezoidal method is a first order method, while the Gauss quadrature method is of higher order when the number of nodes is increased.

#### <span style="color:DodgerBlue"> 14. </span> Mini-project


Explanation of the code :

For the 'gauss_legendre_norm' function, we use the change of variable seen before to compute the integral of the function over $[-1, 1]$ from the original integral over $[0, |a|]$.
The original interval is like this because we use the symmetry of the integral to compute the integral over $[0, |a|]$ and the fact that we know its value over $\mathbb{R}_+$ or $\mathbb{R}_-$.



For the 'gauss_hermite_norm' function, we do not need to change the variable since $\mathcal{X} = \mathbb{R}$. We just need to adjust f to match the weight function : $$\int_{-\infty}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \mathbb{1}(x \leq a)  dx = \int_{-\infty}^{\infty} w(x) \frac{e^{\frac{x^2}{2}}}{\sqrt{2\pi}} \mathbb{1}(x \leq a) dx $$



For the 'gauss_laguerre_norm' function, we just use the symmetry to take advantage of  $\mathcal{X} = \mathbb{R_+}$ like we did with Legende Polynomials. We then again adjust the function to use with the nodes to match the weight function through a change of variable $u = \frac{x^2}{2}$ : $$\int_{0}^{\infty} \frac{e^{\frac{-x^2}{2}}}{\sqrt{2\pi}} \mathbb{1}(x \leq |a|)  dx = \int_{0}^{\infty} w(u) \frac{1}{2 \sqrt{\pi u}} \mathbb{1}(u \leq \frac{a^2}{2}) du $$

```python
# Estimation of the integral

def gauss_legendre_norm(a,n):
    nodes, weights = roots_legendre(n)
    m = abs(a) / 2
    h = m
    I_hat = 0
    for i in range(n) :
        I_hat += (weights[i] * norm.pdf(m + h * nodes[i]))*h
    if a < 0:
        return 1/2 - I_hat
    else:
        return 1/2 + I_hat

    
def gauss_hermite_norm(a,n):
    nodes, weights = roots_hermite(n)
    I_hat = 0
    for i in range(n) :
        I_hat += weights[i] * (nodes[i] <= a) * np.exp((nodes[i]**2)/2) / np.sqrt(2*np.pi)
    return I_hat

def gauss_laguerre_norm(a,n):
    nodes, weights = roots_laguerre(n)
    I_hat = 0
    for i in range(n) :
        I_hat += weights[i] * (nodes[i] <= (a**2)/2) /(2*np.sqrt(np.pi*nodes[i]))
    if a < 0:
        return 1/2 - I_hat
    else:
        return 1/2 + I_hat
```

```python
# Legendre vs Hermite vs Laguerre

n = np.arange(10, 100)

# Store the errors
legendre = []
hermite = []
laguerre = []
a = np.random.uniform(-1000, 1000)
# exact value of the probability
exact = norm.cdf(a)
for k in n:
    legendre.append(abs(gauss_legendre_norm(a, k) - exact))
    hermite.append(abs(gauss_hermite_norm(a, k) - exact))
    laguerre.append(abs(gauss_laguerre_norm(a, k) - exact))
    
    
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))

ax1.plot(n, legendre, label='Legendre', color='red')
ax1.set_ylabel('Absolute Error')

ax1.set_xlabel('Degree of Polynomial')
ax1.legend()

ax2.plot(n, hermite, label='Hermite', color='green')
ax2.set_ylabel('Absolute Error')
ax2.set_xlabel('Degree of Polynomial')
ax2.legend()

ax3.plot(n, laguerre, label='Laguerre', color='blue')
ax3.set_ylabel('Absolute Error')
ax3.set_xlabel('Degree of Polynomial')
ax3.legend()

plt.show()

```

The Hermite methods clearly works better than the Laguerre methods. This is because the Hermite polynomials are computed over the weight function $e^{-x^2}$, which is very close to the density function of the normal law. Thus the nodes of the Hermite polynomials are closer to the nodes of the normal law, and the quadrature formula is more accurate.

## <span style="color:green"> Part III : Richardson </span>

#### <span style="color:DodgerBlue"> 14. </span> Show that the expression given removes the first $j$ error terms using $i$ nodes.

We will prove by induction that $A_{i,j}=\frac{A_{i,j-1}-\alpha^jA_{i-1,j-1}}{1-\alpha}$ removes the first $j$ error terms using $i$ nodes.

Let's show by induction on $j$ that $\forall j$, there exists $(b_k)$ such that $\forall i$, $A_{i,j} = A + \sum_{k=j+1} b_k \alpha^{ik} h^k$.

Base case : j = 0. Directly with the Taylor expansion of A, we have

$$A_{i,0} = A + \sum_{k=1} b_k \alpha^{ij} h^j$$

Inductive Hypothesis : Take $j \geq 0$ such that the property is verified for all $k \leq j$.

Let's show the property remains true for j+1.

Take $i \in \mathbb{N}$.

By definition, we have : $$A_{i,j+1} = \frac{A_{i,j}-\alpha^{j+1}A_{i-1,j}}{1-\alpha^{j+1}}$$

Using the hypothesis of induction, we have : $$A_{i,j+1} = \frac{A + \sum_{k=j+1} b_k \alpha^{ik} h^k - \alpha^{j+1} (A + \sum_{k=j+1} b_k \alpha^{(i-1)k} h^k)}{1-\alpha^{j+1}}$$

Hence, $$A_{i,j+1} = A + \frac{\sum_{k=j+2} (b_k \alpha^{ik} h^k - \alpha^{j+1} b_k \alpha^{(i-1)k} h^k)}{1-\alpha^{j+1}} + \frac{b_{j+1} \alpha^{i(j+1)} h^{j+1} - \alpha^{j+1} b_{j+1} \alpha^{(i-1)(j+1)} h^{j+1}}{1-\alpha^{j+1}} $$

Thus, $$A_{i,j+1} = A + \sum_{k=j+2} (b_k - \alpha^{j+1} b_k \alpha^{-k}) \alpha^{ik} h^k + \frac{(b_{j+1} h^{j+1}) (\alpha^{i(j+1)} - \alpha^{j+1} \alpha^{(i-1)(j+1)})}{1-\alpha^{j+1}} $$

Finally, since $\alpha^{i(j+1)} - \alpha^{j+1} \alpha^{(i-1)(j+1)} = \alpha^{i(j+1)} - \alpha^{i(j+1)} = 0$   with $c_k = b_k - \alpha^{j+1} b_k \alpha^{-k}$ we have : $$A_{i,j+1} = A + \sum_{k=j+2} c_k \alpha^{ik} h^k$$

Therefore, the induction is finished.

Conclusion : the property is verified for all $j \geq 0$ and all $i \in \mathbb{N}$. Thus, for all $j \geq 0$ $A_{i,j}$ removes the first $j$ error terms. (The sum begins at $j+1$).

And, finally, by construction (the order i goes down by one at each iteration), $A_{i,j}$ is computed through i nodes.

#### <span style="color:DodgerBlue"> 15. </span> Implement Richardson's acceleration on the trapezoidal rule.


```python
#Richardson acceleration on the trapezoidal rule

def richardson(f, a, b, M, n):
    A = np.zeros((n, n))  # array to compute the A_{i,j} values
    A[0, 0] = trapezoidal(f, a, b, M)
    for i in range(1, n):
        M = 2*M
        A[i, 0] = trapezoidal(f, a, b, M)
        alpha = 1/4  # alpha is 1/4 and not 1/2 because we are using the trapezoidal rule so the powers are multiples of 2
        for j in range(1, i+1):
            A[i,j] = (A[i,j-1] - alpha * A[i-1,j-1]) / (1 - alpha)
            alpha = alpha/4
    return A[n-1,n-1]
```

```python
# Compute the integration error for the trapezoidal rule with and without Richardson's acceleration and in comparison
h_values = np.logspace(-1, -5, num=50)
trapezoidal_errors = np.zeros_like(h_values)
romberg_errors = np.zeros_like(h_values)
quadrature = np.zeros_like(h_values)
nodes, weights = roots_legendre(n = 100)
exact = quad(f1, 0, 1)[0]

for i, h in enumerate(h_values):
    trapezoidal_integral = trapezoidal(f1, 0, 1, int(1/h))
    romberg_integral = richardson(f1, 0, 1, int(1/h), 10)
    trapezoidal_errors[i] = abs(trapezoidal_integral - exact)
    romberg_errors[i] = abs(romberg_integral - exact)
    quadrature[i] = abs(gauss_legendre_simple(f1,0,1,nodes,weights) - exact)

# Plot the integration errors as a function of 1/h
fig, ax = plt.subplots()
ax.loglog(1/h_values, trapezoidal_errors, label="Trapezoidal Rule")
ax.loglog(1/h_values, romberg_errors, label="Romberg's Integration")
ax.loglog(1/h_values, quadrature, label="Gauss-Legendre Quadrature")
ax.set_xlabel("1/h")
ax.set_ylabel("Integration Error")
ax.legend()
plt.show()
```

When 1/h is sufficiently low, Romberg's Integration is better than the gauss quadrature. Furthermore we see that the acceleration works clearly better than the trapezoidal rule without the acceleration.
